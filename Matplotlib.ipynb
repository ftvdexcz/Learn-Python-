## Pyplot API vs Object-Oriented API 
* Pyplot API -> Quickly 
* OO API -> Advanced 

### Pyplot API 

```python
x = np.linspace(0,10,1000) # Chia x thành 1000 điểm trong khoảng từ 0->10 
plt.plot(x,np.sin(x),color="red",label="sin(x)")
plt.plot(x,np.cos(x),color="green",label="cos(x)")
plt.xlabel("Truc hoanh") 
plt.ylabel("Truc tung")
#plt.xlim([0,8])
#plt.ylim([-1,1])

#plt.axis([0,10,-1,1]) # giống xlim ylim 
plt.axis("equal") # đồ thị cân đối với hình 

plt.legend() # Chú thích cho đồ thị, thêm label="" ở từng đồ thị để chú thích 
plt.title("Sin and Cos") # Title đồ thị
plt.show()
```

### Object-Oriented API 

```python
#1. Prepare data
X = [1,2,3,4]
y = [11,22,33,44]

#2. Setup plot 
fig, ax = plt.subplots(figsize=(7,7)) #Figure size = (width, height)

#3. Plot the data
ax.plot(X,y)

ax.set(title="A simple plot",
       xlabel="x-axis",
       ylabel="y-axis")
```

## Most common types of Matplotlib plots 
* line 
* scatter
* bar 
* histogram 
* subplots()

### Line (Liên tục)

```python
# Create array 
X = np.linspace(0,10,100)

## OO API 
fig, ax = plt.subplots(figsize=(7,7))
ax.plot(X,X**3,color="red")
```

### Scatter  (Biểu đồ phân tán)

```python
# pyplot API 
plt.scatter(X,np.exp(X)) # y = e^x 
```

```python
## OO API 

rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.randn(100)
fig,ax = plt.subplots(figsize=(7,7))
ax.scatter(x,y,c=colors,cmap='viridis',alpha=0.3)
```

### Bar

#### Vertical

```python
#Prepare data from a python dictionary 
soft_drink_prices = {
    "Coke": 10,
    "Pepsi": 12,
    "Sprite": 8
}
fig, ax = plt.subplots(figsize=(7,7))
ax.bar(soft_drink_prices.keys(),soft_drink_prices.values())
ax.set(title="Soft Drink Prices", 
       ylabel="Price ($)")
```

#### Horizontal


```python
fig, ax = plt.subplots(figsize=(7,7))
ax.barh(list(soft_drink_prices.keys()),list(soft_drink_prices.values()))
```

### Histogram 

```python
np.random.seed(42)
student_height = np.random.normal(170, 10, 250)
# pyplot API
plt.hist(student_height,bins=30);
```

### Subplots

```python
# Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x,x/2)
ax1.plot(x,x*2)

ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(soft_drink_prices.keys(),soft_drink_prices.values());
ax4.hist(student_height,bins=30);
```

```python
# Option 2
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,x/2)
ax[0,0].plot(x,x*2)

ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(soft_drink_prices.keys(),soft_drink_prices.values());
ax[1,1].hist(student_height,bins=30);
```
