## Data Manipulation 

```python
df.item_price.dtype 
```

```python
# Apply() function
df.item_price = df.item_price.apply(lambda x: float(x.replace('$',''))) #bỏ dấu $ 
```

```python
df.head()
```

```python
# Tạo column mới tên là total_price
df["total_price"] = df["quantity"] * df["item_price"]
```

```python
df.head(5)
```

### How much was the revenue for the period in the dataset

```python
revenue = df["total_price"].sum() #tổng doanh thu 
print(revenue)
```

### Which was the most-ordered item ? 

```python
# group by 
c = df.groupby("item_name")["quantity"].sum()
c.sort_values()
```
